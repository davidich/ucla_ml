```python
import numpy as np

float_formatter = "{:.2f}".format
float_formatter(1.234567E-3)
np.set_printoptions(formatter={'float_kind':float_formatter})
```

### Generate test data

```python
import matplotlib.pyplot as plt

def generate_data(): 
    """
    Generate movie feedbacks.
    Dataset contains 2 features: user_score (x1) and critic_score (x2).
    If a user doesn't like a movie than y=0, otherwise y=1.
    Function assumes that a user likes a movie if its total_score >= 6.
    total_score = critic_score * 2 + user_score * 2
    """
    np.random.seed(74837)

    # Random user and critic scores
    n = 100
    user_score = 5*np.random.random(size=n)
    critic_score = 5*np.random.random(size=n)
    X = np.concatenate((user_score.reshape(len(user_score), 1),
                        critic_score.reshape(len(critic_score), 1)),
                        axis=1)
    
    # Set y = 1 above decision boundary
    boundary = lambda x1, x2: x1 + 2*x2 - 6
    y = np.zeros(n)
    for i in range(n):
        if boundary(user_score[i], critic_score[i]) > 0:
            y[i] = 1
            
    return X, y, user_score, critic_score

X, y, user_score, critic_score = generate_data()

# Create figure
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111)

# Colors for scatter points
colors = ['red', 'green']
colors_data = [colors[int(i)] for i in y]

# Plot sample data
ax.scatter(user_score, critic_score, color=colors_data, s=100, edgecolor='black', linewidth=2, alpha=0.7)
ax.grid(linestyle='--', linewidth=2, color='gray', alpha=0.2)
ax.set(xlabel='User Score', ylabel='Critic Score')
plt.show()
```

### Train model

```python
from logistic_regression import MyLogisticRegression

X, y, _, _ = generate_data()
mylr = MyLogisticRegression(learning_rate=0.05, max_iterations=200000)
mylr.fit(X, y, plot_cost=True)
```

### Visualize model

```python
import numpy as np

# Create figure
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111)

# Colors for scatter points
colors = ['red', 'green']
colors_data = [colors[int(i)] for i in y]

# Plot decision boundary
n = 50
x1_bound, x2_bound = np.meshgrid(np.linspace(0, 5, n), np.linspace(0, 5, n))
sigmoid = lambda t: 1/(1 + np.exp(-t))
prob = sigmoid(mylr.intercept_ + x1_bound*mylr.coef_[0][0] + x2_bound*mylr.coef_[0][1])
ax.contour(x1_bound, x2_bound, prob, levels=[0.5], colors='black')

# Plot sample data
ax.scatter(user_score, critic_score, color=colors_data, s=100, edgecolor='black', linewidth=2, alpha=0.7)
ax.grid(linestyle='--', linewidth=2, color='gray', alpha=0.2)
ax.set(xlabel='x', ylabel='y')

plt.show()
```
